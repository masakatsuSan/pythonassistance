```python
## linear search algorithm

def indexfinding(arr,k):
    for i in range(len(arr)):
        if arr[i] == k :
            return i 
result = indexfinding([1,43,6,34,7,67,34,22],7)
print(result)
```

```python
## Binary search algorithm

def binarySearch(arr,x,i,j):
    while i<= j :
        mid = i + (j-1)//2
        if arr[mid] == x :
            return mid
        elif arr[mid]< x :
            i = mid + 1 
        else :
            j = mid -1  

    return -1 
arr = [10,20,30,40,50,60,70]
i = 0
j = len(arr)-1
print(binarySearch(arr,20,i,j))
```

```python
## recursive binary method

def binarySearch(arr,x,i,j):
    while i<= j :
        mid = i + (j-1)//2
        if arr[mid] == x :
            return mid
        elif arr[mid]< x :
            return binarySearch(arr,x,mid+1,j) 
        else :
            return binarySearch(arr,x,i,mid-1)

    return -1 
arr = [10,20,30,40,50,60,70]
i = 0
j = len(arr)-1
print(binarySearch(arr,20,i,j))
```
