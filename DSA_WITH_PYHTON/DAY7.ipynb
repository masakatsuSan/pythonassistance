```python
def bubblesort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0,n-1-i): # here we are excluding last numbers , then second last ,.....
            if arr[j] > arr[j+1]:
                arr[j],arr[j+1] = arr[j+1],arr[j]
    return arr   
arr = [20,30,35,10,90,75] #len = 6
print(bubblesort(arr))
## Bubble sort ## time complexity --> O(n^2)
```
